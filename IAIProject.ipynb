```python
import rasterio

def extract_values(df, raster_path, col):
    coords = [(x, y) for x, y in zip(df['Longitude'], df['Latitude'])]

    with rasterio.open(raster_path) as src:
        values = [val[0] for val in src.sample(coords)]

    df[col] = values
    return df
```

```python
path = r"C:\IAI Project\raw data"

# 2019
df_2019 = extract_values(df_2019, path + r"\NDWI_2019.tif", "NDWI")
df_2019 = extract_values(df_2019, path + r"\LST_2019.tif", "LST")
df_2019 = extract_values(df_2019, path + r"\RAIN_2019.tif", "Rain")

# 2021
df_2021 = extract_values(df_2021, path + r"\NDWI_2021.tif", "NDWI")
df_2021 = extract_values(df_2021, path + r"\LST_2021.tif", "LST")
df_2021 = extract_values(df_2021, path + r"\RAIN_2021.tif", "Rain")

# 2022
df_2022 = extract_values(df_2022, path + r"\NDWI_2022.tif", "NDWI")
df_2022 = extract_values(df_2022, path + r"\LST_2022.tif", "LST")
df_2022 = extract_values(df_2022, path + r"\RAIN_2022.tif", "Rain")
```

```python
final_df = pd.concat([df_2019, df_2021, df_2022], ignore_index=True)
```

```python
final_df['wet_heat'] = final_df['NDWI'] * final_df['LST']

final_df['is_night'] = (
    (final_df['Hour'] < 6) | (final_df['Hour'] > 20)
).astype(int)
```

```python
final_df['risk'] = final_df['Accident_Severity'].apply(
    lambda x: 1 if x <= 2 else 0
)
```
